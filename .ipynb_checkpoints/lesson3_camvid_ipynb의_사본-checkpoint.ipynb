## Datasets

```python
size = src_size//2
free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=8
else:           bs=4
print(f"using bs={bs}, have {free}MB of GPU RAM free")
```

```python
src = (SegmentationItemList.from_folder(path_img)
       .split_by_fname_file('../valid.txt')
       .label_from_func(get_y_fn, classes=codes)) #datasets 만들기
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats)) #정규화
```

```python
data.show_batch(2, figsize=(10,7))
```

```python
data.show_batch(2, figsize=(10,7), ds_type=DatasetType.Valid) #데이터 조각을 결합해 사진에 색상 지정
```

## Model

```python
name2id = {v:k for k,v in enumerate(codes)}
void_code = name2id['Void']

def acc_camvid(input, target):  #camvid
    target = target.squeeze(1)
    mask = target != void_code
    return (input.argmax(dim=1)[mask]==target[mask]).float().mean()
```

```python
metrics=acc_camvid #metrics=accuracy
```

```python
wd=1e-2 
#강의자료에서는 Ir(learning rate)인데 왜 여기서는 wd(weight decay)를 사용하나요?
#여기서부터 go big 전까지의 코드가 강의자료와 약간 다릅니다...
```

> ## 매우 좋은 질문입니다. 
> ## 강의에 비해 코드가 더 개선되어 일부 다른 내용이 있습니다.

[가중치 감쇠](https://ko.d2l.ai/chapter_deep-learning-basics/weight-decay.html)

[Ir과 wd의 차이?](https://stats.stackexchange.com/questions/29130/difference-between-neural-net-weight-decay-and-learning-rate)

몇 개 검색해봤는데도 잘 이해가 되질 않습니다...

> ## lr은 training 할 때, fitting하며 Gradient descent 과정에서 weight를 변화할 때 얼마나 더 변화시켜볼지를 결정하는 값이죠.
> ## wd는 training 이후 이를 다른 데이터들에 적용해보며 그것을 활용하거나, test하는데 그 때 발생할 수 있는 error를 줄이기 위해(이를 generalization라고 합니다.) 정규화 과정에 쓰이는 값입니다.

> ## 정규화란 무엇인지 한 번 찾아보면 도움이 될 것 같아요

```python
learn = unet_learner(data, models.resnet34, metrics=metrics, wd=wd)
```

```python
lr_find(learn)
learn.recorder.plot() #그래프 개형이 조금 다르네요...?
```

```python
lr=3e-3
```

```python
learn.fit_one_cycle(10, slice(lr),pct_start=0.9) 
#pct_start:Percentage of total number of epochs when learning rate rises during one cycle.
```

```python
learn.save('stage-1')
```

```python
learn.load('stage-1');
```

```python
learn.unfreeze()
```

```python
lr_find(learn)
learn.recorder.plot()
```

```python
lrs = slice(lr/400,lr/4)
```

```python
learn.fit_one_cycle(12, lrs,pct_start=0.8)
```

```python
learn.save('stage-2');
```

## Go big

You may have to restart your kernel and come back to this stage if you run out of memory, and may also need to decrease `bs`.

```python
learn.destroy()

size = src_size

free = gpu_mem_get_free_no_cache()
# the max size of bs depends on the available GPU RAM
if free > 8200: bs=3 #강의자료에서는 batch size가 4
else:           bs=1
print(f"using bs={bs}, have {free}MB of GPU RAM free") #모든 픽셀을 사용하기 위해 bs 줄임.
```

```python
data = (src.transform(get_transforms(), size=size, tfm_y=True)
        .databunch(bs=bs)
        .normalize(imagenet_stats)) #정규화
```

```python
learn = unet_learner(data, models.resnet34, metrics=metrics)
```

```python
learn.load('stage-2');
```

```python
lr_find(learn)
learn.recorder.plot()
```

```python
lr=1e-3
```

```python
learn.fit_one_cycle(10, slice(lr),pct_start=0.8)
```

```python
learn.save('stage-1-big')
```

```python
learn.load('stage-1-big');
```

```python
learn.unfreeze()
```

```python
lrs = slice(1e-6,lr/10)
```

```python
learn.fit_one_cycle(10, lrs)
```

```python
learn.save('stage-2-big')
```

```python
learn.load('stage-2-big');
```
